```python

def clean_house_price(df):
    clean_df = df.iloc[:,[0,1,4,5,6]]
    clean_df['Year'] = clean_df.columns[3][-4:]
    clean_df['Quarter'] = "Q" + str(clean_df.columns[3][-7:-6])
    clean_df.columns = ['city', 'suburb', 'sales', 'median', 'median_change', 'year', 'quarter']
    clean_df.head()
    return clean_df
```

```python
sa_house_price = pd.DataFrame()
for file_name in dir_list:
    print(file_name)
    df = pd.read_excel(file_path + file_name, engine='openpyxl')
    clean_df = clean_house_price(df)
    sa_house_price = pd.concat([sa_house_price, clean_df])
```
